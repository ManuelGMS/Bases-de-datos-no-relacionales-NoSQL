```python
# **** LIBRERIAS ******************************************************************************************* 

from pprint import pprint; # Lireria Pretty Print.
from datetime import datetime; # Libreria para trabajar con fechas y horas.
from pymongo import MongoClient; # Linreria del cliente MongoDB. 

# **** CONEXION A LA BBDD **********************************************************************************

# Obtenermos un objeto cliente para conectarnos a la BBDD.
cliente = MongoClient(); # Usa la conexion por defecto.
# cliente = MongoClient('mongodb://localhost:27017'); # Esta es la conexion por defecto.

db = cliente.test; # Indicamos la BBDD a la que queremos conectarnos con el cliente.
# db = cliente["test"]; # Indicamos la BBDD a la que queremos conectarnos con el cliente.

# **** INSTRUCCIONES SOBRE LA BBDD *************************************************************************

# Mostrar tablas y cantidad de documentos.
for col in db.list_collection_names():
    
    print(col,db[col].count_documents({}));

# DROP TABLE <TableName> IF EXISTS;
if("Empleados" in db.list_collection_names()): # Ver si la coleccion existe.

    db.Empleados.drop(); # Borra la coleccion.

# **** INSTRUCCIONES SOBRE LAS COLECCIONES DE LA BBDD ******************************************************

# WRITE:

one = db.Empleados.insert_one(
    { 
        "nombre": "Alfred" , 
        "departamento": "Ing-Tec" , 
        "jefe": True ,
        "sueldo": 4000
    }
);
many = db.Empleados.insert_many([ 
    { "nombre": "Carlos" , "ocupacion": "programador" } , 
    { "nombre": "Sergio" , "ocupacion": "analista" , "Inicio laboral": datetime.utcnow() } ,
    { "nombre": "Estrella" , "ocupacion": "programador" }
]);
print("Insertado:",one.inserted_id);
print("Insertado:",many.inserted_ids);

# UPDATE:

db.Empleados.update_one(
    { "nombre": "Alfred" }
    ,
    { 
        "$set": { "jefe": False , "ocupacion": "Arquitecto Software" } # EStablece o crea el campo.
        ,
        "$inc": { "sueldo": -1000 } # Incrementa / Decrementa el valor del campo.
    }
);

updated = db.Empleados.update_many(
    { "ocupacion": "programador" }
    ,
    { "$set": { "sueldo": 1850 } }
);
print("Updated:",updated.modified_count);

upserted = db.Empleados.update_many(
    { "nombre": "Gilberto" , "ocupacion": "testeador" , "sueldo": 2000 }
    ,
    { "$inc": { "sueldo": -150 } }
    ,
    upsert=True
);
print("Upserted:",upserted.upserted_id);
    

# READ:

# ReadOne
# register = db.Empleados.find_one({},{});
# print(register);

# ReadAll
cursor = db.Empleados.find({},{ "_id": False });
print("Encontrados:",db.Empleados.count_documents({}));
for row in cursor:
    pprint(row); # print(row["collumnName"]);
      
# DELETE

dOne = db.Empleados.delete_one({ "nombre": {"$eq":"Sergio"} });
print("Borrado:",dOne.deleted_count);

dMany = db.Empleados.delete_many({ "nombre": {"$eq":"Carlos"} });
print("Borrado:",dMany.deleted_count);

dMany = db.Empleados.delete_many({ "nombre": {"$eq":"Alfred"} });
print("Borrado:",dMany.deleted_count);
```
